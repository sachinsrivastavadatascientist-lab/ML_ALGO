```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn import datasets
data = datasets.load_iris()
X = data.data
y = data.target
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

Applying PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2) # to visualize 2 dimension
pca_scaled_data =pca.fit_transform(x_scaled)
```

```python
pca_scaled_data
```

```python
plt.scatter(X[:,0],X[:,1],c = y)
```

# Algomerative Clustering

```python
import scipy.cluster.hierarchy as sc

plt.figure(figsize = (20,7))
plt.title("dendogram")

sc.dendrogram(sc.linkage(pca_scaled_data,method = "ward"))
plt.xlabel("sample_data")
plt.ylabel;("Euclidean distabnce")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 2,linkage = "ward")
cluster.fit(pca_scaled_data)
```

```python
cluster.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c = cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coeff = []
for k in range(2,10):
    cluster = AgglomerativeClustering(n_clusters = k,linkage = "ward")
    cluster.fit(X)
    score = silhouette_score(X,cluster.labels_)
    silhouette_coeff.append(score)
```

```python
silhouette_coeff
```

```python
plt.plot(range(2,10),silhouette_coeff)
plt.xticks(range(2,10))
plt.xlabel("No of clusters")
plt.ylabel("silhotte score")
```
